```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
train_data = pd.read_csv("train_titanic.csv")
test_data = pd.read_csv("test _titanic.csv")

test_data
```

```python
train_data["CryoSleep"]=train_data["CryoSleep"].replace([False,True],[0,1])
train_data["VIP"]=train_data["VIP"].replace([False,True],[0,1])
train_data["Transported"]=train_data["Transported"].replace([False,True],[0,1])
test_data["CryoSleep"]=test_data["CryoSleep"].replace([False,True],[0,1])
test_data["VIP"]=test_data["VIP"].replace([False,True],[0,1])

train_data
```

```python


enc = LabelEncoder()
```

```python
train_data["ShoppingMall"] = train_data["ShoppingMall"].fillna(train_data["ShoppingMall"].mean())
train_data["FoodCourt"] = train_data["FoodCourt"].fillna(train_data["FoodCourt"].mean())
train_data["RoomService"] = train_data["RoomService"].fillna(train_data["RoomService"].mean())
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].mean())
train_data["VRDeck"] = train_data["VRDeck"].fillna(train_data["VRDeck"].mean())
train_data["Spa"] = train_data["Spa"].fillna(train_data["Spa"].mean())

train_data["HomePlanet"] = enc.fit_transform(train_data["HomePlanet"])
train_data["Destination"] = enc.fit_transform(train_data["Destination"])
train_data = train_data.drop(["Name"],axis=1)


test_data["ShoppingMall"] = test_data["ShoppingMall"].fillna(test_data["ShoppingMall"].mean())
test_data["FoodCourt"] = test_data["FoodCourt"].fillna(test_data["FoodCourt"].mean())
test_data["RoomService"] = test_data["RoomService"].fillna(test_data["RoomService"].mean())
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].mean())
test_data["VRDeck"] = test_data["VRDeck"].fillna(test_data["VRDeck"].mean())
test_data["Spa"] = test_data["Spa"].fillna(test_data["Spa"].mean())


test_data["HomePlanet"] = enc.fit_transform(test_data["HomePlanet"])
test_data["Destination"] = enc.fit_transform(test_data["Destination"])

test_data["Cabin"].fillna(test_data["Cabin"].value_counts().index[0],inplace=True)
test_data["CryoSleep"].fillna(test_data["CryoSleep"].value_counts().index[0],inplace=True)
test_data["Destination"].fillna(test_data["Destination"].value_counts().index[0],inplace=True)
test_data["VIP"].fillna(test_data["VIP"].value_counts().index[0],inplace=True)

test_data = test_data.drop(["Name"],axis=1)

train_data = train_data[(train_data["CryoSleep"]>=0) & (train_data["Cabin"].isna()==False)& (train_data["Destination"].isna()==False)& (train_data["VIP"].isna()==False)]

train_data
```

```python
train_data["cabin_first"] = train_data["Cabin"].str.split("/",expand=True)[0]
train_data["cabin_first"] = enc.fit_transform(train_data["cabin_first"])

test_data["cabin_first"] = test_data["Cabin"].str.split("/",expand=True)[0]
test_data["cabin_first"] = enc.fit_transform(test_data["cabin_first"])

train_data["cabin_second"] = train_data["Cabin"].str.split("/",expand=True)[1]

test_data["cabin_second"] = test_data["Cabin"].str.split("/",expand=True)[1]

train_data["cabin_third"] = train_data["Cabin"].str.split("/",expand=True)[2]
train_data["cabin_third"] = enc.fit_transform(train_data["cabin_third"])

test_data["cabin_third"] = test_data["Cabin"].str.split("/",expand=True)[2]
test_data["cabin_third"] = enc.fit_transform(test_data["cabin_third"])
```

```python
PassengerId = test_data["PassengerId"]
Y_train = train_data["Transported"]
Y_train
```

```python
X_train = train_data.drop(["Cabin","PassengerId","Transported"],axis=1)
test_data.drop(["Cabin","PassengerId"],inplace=True,axis=1)

test_data
```

```python
train_test_data = pd.concat([train_data,test_data],axis=0)
```

```python
model = LogisticRegression( solver='lbfgs')
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(test_data)
y_pred = pd.Series(y_pred)
y_pred = y_pred.replace([0,1],["False","True"])
```
