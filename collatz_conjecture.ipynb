```python
def colatz_sequence(x):
    arr = []
    while True:
        
        if x%2 == 0:
            x/=2
        else:
            x= 3*x + 1
        arr.append(x)
        if x==1:
            break
    return arr

    
```

```python

plt.plot(colatz_sequence(77031),label = "starting number 850")
plt.legend()
plt.savefig('77031.jpg')
```
